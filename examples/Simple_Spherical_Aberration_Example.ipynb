```python
import numpy as np
import matplotlib.pyplot as plt

import tracepy as tp
```

```python
lens = {
    'name': 'lens',
    'action': 'refraction',
    'P': 2.,
    'kappa': -.004,
    'c': -.5,
    'Diam': 2.2
}

back_lens = {
    'name': 'back_lens',
    'action': 'refraction',
    'P': 1.,
    'c':.5,
    'kappa': -.004,
    'N': 1.5,
    'Diam': 2.2
}

stop = {
    'name': 'stop',
    'action': 'stop',
    'P': 3.,
    'Diam': 2.2
}

geo_1 = [back_lens, lens, stop]
ray_group = tp.ray_plane(geo_1, [0., 0., 0.], 1.1, d=[0.,0.,1.], nrays=100)
```

```python
tp.plotyz(geo_1, ray_group)
```

```python
tp.spotdiagram(geo_1, ray_group)
```

```python
ray_group_2 = tp.ray_plane(geo_1, [0., 0., 0.], 1.1, d=[0.,0.,1.], nrays=1000)
tp.rayaberration(geo_1, ray_group_2)
```

```python
vlens = {
    'name': 'stop',
    'vary': ['P']
}

vlens2 = {
    'name': 'back_lens',
    'vary': ['c', 'kappa', 'N']
}

vlens3 = {
    'name': 'lens',
    'vary': ['c', 'kappa']
}

vary_list = [vlens, vlens2, vlens3]
```

```python
geo_opt = tp.optimize(geo_1, vary_list, typeof='least_squares')
ray_group_opt = tp.ray_plane(geo_opt, [0., 0., 0.], 1.1, d=[0.,0.,1.], nrays=100)
```

```python
tp.plotyz(geo_opt, ray_group_opt)
```

```python
tp.spotdiagram(geo_opt, ray_group_opt)
```

```python
ray_group_3 = tp.ray_plane(geo_opt, [0., 0., 0.], 1.1, d=[0.,0.,1.], nrays=1000)
tp.rayaberration(geo_1, ray_group_3)
```
